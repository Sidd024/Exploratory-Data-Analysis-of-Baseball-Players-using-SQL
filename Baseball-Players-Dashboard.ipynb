```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
#loading the entire dataset
baseball_dataset = pd.read_csv("baseball_data.csv");
```

```python
#top 6 players having max heights
top_max_height = baseball_dataset.nlargest(6,'height')[['name','height']]
plt.figure(figsize=(16,5))
plt.title("Top 6 players having max heights")
plt.bar(top_max_height['name'],top_max_height['height'])
plt.ylabel("Player their heights")
plt.xlabel("Top 6 players showing their name")
plt.grid()
plt.show()
```

```python
#top 6 players having max weights
top_max_weight = baseball_dataset.nlargest(6,'weight')[['name','weight']]
plt.figure(figsize=(16,5))
plt.title("Top 6 players having max weights")
plt.plot(top_max_weight['name'],top_max_weight['weight'],marker='o')
plt.xlabel('Top 6 players showing their name')
plt.ylabel('Top 6 players showing their weights')
plt.show()
```

```python
#top 10 HR(Home Run) of Players
top_max_HR = baseball_dataset.nlargest(10, 'HR')[['name', 'HR']]
plt.figure(figsize=(16,5))
plt.title("Top 10 players having max home run")
plt.pie(list(top_max_HR['HR']),labels=list(top_max_HR['name']),autopct='%1.1f%%')
plt.show()
```

```python
baseball_count = (baseball_dataset['handedness'] .str.split(',').explode().str.strip().value_counts().head(10))
plt.figure(figsize=(16,5))
plt.plot(baseball_count.index,baseball_count.values, marker='o')
plt.title('Separate Handedness Representation')
plt.xlabel('Total Handedness Label Name')
plt.ylabel('Total Handedness Count')
plt.show()
```
